**Question-2:** Assume that two *binary* strings are given: [00110] and [11001]. 
* Convert these strings to decimal numbers.
* (*Optional*) If these binary strings represent floating numbers between 1.5 (**L**ower **B**ound) and 7.3 (**U**pper **B**ound), please calculate their values.
* Writing a function is preferable (e.g. *binary2real(binary_string, LB, UB)*).
* (*Optional*) Function () can be written in a general form which works for any length of binary string and LB-UB pair provided as inputs to the function.

**Hint:** 
* The first binary string [00110] equals to 6, how? Array multiplication and summation operations are needed (*Google* for "How to Convert from Binary to Decimal"). 
* (*Optional*) In order to convert these binary strings to decimals, you first need to decode them into decimals, and then perform the scaling. For instance, [00000] will be decoded as 0 and will represent the lower bound, i.e. 1.5. Likewise, [11111] will be decoded as 31 and will represent the upper bound, i.e. 7.3 in this case. Then it is matter of scaling to find out what [00110] represents. If the scaling is confusing, then you can directly use the following formula:

$$x_i=Lower\:bound + \left(\frac{Upper\:bound-Lower\:bound}{2^{Length\:of\:the\:binary\:string}-1 }\right).decoded\:value, $$
which would result as
$$x_i=1.5 + \left(\frac{7.3-1.5}{2^5-1 }\right).6 = 2.6226 $$

```python
def binaryToDecimal(binary_str):
    result = 0; 
    binVal = pow(2, len(binary_str) -1)
    for x in range(0, len(binary_str)):
        if(binary_str[x] == '1'):
            result += binVal
        binVal /= 2
    return result

#test cases for binaryToDecimal: 

binaryToDecimal("00110") # 6
binaryToDecimal("11001") # 25
binaryToDecimal("11111") # 31

#Optional function binary2real

def binary2real(binary_str, LB, UB): 
    decodedVal = binaryToDecimal(binary_str)
    result = LB + (((UB - LB) / (pow(2, len(binary_str)) - 1) * decodedVal))
    return result

#test cases for binary2real: 
binary2real("00110", 1.5, 7.3) # 2.6226
binary2real("11111", 1.5, 7.3) # 7.3
binary2real("11001", 1.5, 7.3) # 6.1774     
```
